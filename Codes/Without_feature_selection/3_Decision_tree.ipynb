```python
import pandas as pd 
df=pd.read_csv('D:\Internships\IITK Oswald sir\IVF\Codes\Dataset\cleaned.csv')
df.info()

from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
df
```

```python
dfy = df['Live Birth Occurrence']
x = df.drop('Live Birth Occurrence', axis='columns')

x.info()
```

```python
dfx = scaler.fit_transform(x)
dfx
dfy
```

```python
from sklearn.model_selection import train_test_split

from sklearn.tree import DecisionTreeClassifier

X_train, X_test, y_train, y_test = train_test_split(dfx, dfy, test_size=0.25)
```

```python
model = DecisionTreeClassifier()
params = {
    'criterion': ['gini', 'entropy'],
    'splitter': ['best', 'random'],
    'max_depth': [1, 10, 100, 1000]
}

clf = GridSearchCV(
    estimator=model, 
    param_grid=params, 
    cv=10,  # 10-fold cross validation
    n_jobs=-1  # run in parallel
)
clf.fit(X_train, y_train)

cv_results = pd.DataFrame(clf.cv_results_)
```

```python
cv_results
```

```python
cv_results = cv_results[['mean_test_score', 'param_criterion', 'param_splitter', 'param_max_depth']]
cv_results.sort_values(by='mean_test_score', ascending=False)
```
